```python
#drop highly correlated colums
#drop highly correlated colums
data = data.drop(columns=['Temp9am','Temp3pm','MinTemp'])

```

```python
data.head()
```

```python
data.shape
```

```python
#now handle class imbalance
plt.Figure(figsize=(6,4))
sns.countplot(x="RainToday", data = data)
plt.title("Distribution of RainToday")
plt.show()
```

```python
data["RainToday"].value_counts()
```

```python
#so the majority class is 0 and minority class is 1 now separate data


from sklearn.utils import resample

no = data[data.RainToday == 0]
yes = data[data.RainToday == 1]
yes_oversampled = resample(yes, replace=True, n_samples=len(no), random_state=123)
oversampled = pd.concat([no, yes_oversampled])

fig = plt.figure(figsize = (8,5))
oversampled.RainToday.value_counts(normalize = True).plot(kind='bar', color= ['skyblue','navy'], alpha = 0.9, rot=0)
plt.title('RainToday Indicator No(0) and Yes(1) after Oversampling (Balanced Dataset)')
plt.show()
```

```python
#Split featres and target as x and y 
x = data.drop(columns=['RainToday'])
y = data['RainToday']
```

```python
print(x)
```

```python
print(y)
```

```python
#split the data into train and test data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
len(x_test)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,3))
sns.scatterplot(x='Rainfall',y='RainToday', data=data)
plt.show()
```

```python
numeric_cols = data.select_dtypes(include='number').columns

plt.figure(figsize=(5, 3 * len(numeric_cols)))

for i, column in enumerate(numeric_cols):
    plt.subplot(len(numeric_cols), 1, i + 1)
    sns.histplot(data[column], kde=True)
    plt.title(f"Distribution of {column}")

plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=42)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
#Confusion matrix
y_predict = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
cm
```

```python
%matplotlib inline
plt.Figure(figsize=(10,7))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('real')
```
